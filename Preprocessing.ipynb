```python
import numpy as np
import pandas as pd
import ast
import joblib
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
```

```python
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
links = pd.read_csv('links.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
links.head()
```

```python
new_db = credits.merge(links,on='tmdbId')
```

```python
new_db.shape
```

```python
new_db.head()
```

```python
db = new_db.merge(movies,on='title')
```

```python
db = db[['imdbId','title','overview','genres','keywords','cast','crew']]
```

```python
db['id']=db['imdbId']
```

```python
db = db.drop('imdbId',axis=1)
```

```python
db.head()
```

```python
db.dropna(inplace=True)
```

```python
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
        
    return l
        
```

```python
db['genres'] = db['genres'].apply(convert)
```

```python
db['keywords'] = db['keywords'].apply(convert)
```

```python
db.head()
```

```python
def convert2(obj):
    l = []
    count = 0 
    for i in ast.literal_eval(obj):
        if count!= 3:
            l.append(i['name'])
            count=count+1
        else :
            break
        
    return l
```

```python
db['cast'] = db['cast'].apply(convert2)
```

```python
db.head()
```

```python
def convert3(obj):
    l = []
    count = 0 
    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            l.append(i['name'])
            count=count+1

        
    return l
```

```python
db['crew'] = db['crew'].apply(convert3)
```

```python
db.head()
```

```python
db['overview'] = db['overview'].apply(lambda x:x.split())
```

```python
db.head()
```

```python
db['genres'] = db['genres'].apply(lambda x:[i.replace(" ","") for i in x])
db['keywords'] = db['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
db['cast'] = db['cast'].apply(lambda x:[i.replace(" ","") for i in x])
db['crew'] = db['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
db.head()
```

```python
db['tags'] = db['overview'] + db['genres'] + db['keywords'] + db['cast'] + db['crew']
```

```python
db = db[['id','title','tags']]
```

```python
db.head()
```

```python
db['tags'] = db['tags'].apply(lambda x:" ".join(x))
```

```python
db['tags'] = db['tags'].apply(lambda x:x.lower())
```

```python
db.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
```

```python
stemmer = PorterStemmer()
stwords = stopwords.words('english')
```

```python
def stemming(text):
    y = []
    for i in text.split():
        if i not in stwords:
            y.append(stemmer.stem(i))
    return " ".join(y)
```

```python
db['tags'] = db['tags'].apply(stemming)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000)
```

```python
vector = cv.fit_transform(db['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    index = db[db['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(db.iloc[i[0]].title)
```

```python
recommend('The Dark Knight Rises')
```

```python
joblib.dump(db, 'movie_list.joblib')
```

```python
joblib.dump(similarity, 'similarity.joblib')
```
